```python
import math
import matplotlib.pyplot as plt

# Координаты центров
centers = {
'Центр1': (0, 8),
    'Центр2': (11, 3),
    'Центр3': (14, 9),
    'Центр4': (0, 0),
}
# Координаты точек
points = {
    'A': (8, 13),
    'B': (0, 5),
    'C': (9, 3),
    'D': (9, 11),
    'E': (3, 2),
    'F': (12, 5),
    'G': (3, 4),
    'H': (8, 15),
    'I': (0, 6),
    'J': (6, 0),
    'K': (14, 1),
    'L': (14, 12),
    'M': (15, 5),
    'N': (4, 0),
    'O': (12, 11)
}
```

```python
# Функция для вычисления расстояния
def distance(point, center):
    x_diff = point[0] - center[0]
    y_diff = point[1] - center[1]
    return math.sqrt(x_diff**2 + y_diff**2)
```

```python
# Вычисление расстояний до центров для каждой точки
def get_distances():
    distances = {}
    for point, coordinates in points.items():
        distances[point] = {}
        for center, center_coordinates in centers.items():
            distances[point][center] = distance(coordinates, center_coordinates)
    return distances
```

```python
# Функция для вычисления новых координат центров
def calculate_new_centers():
    distances = get_distances()
    new_centers = {}
    for center, center_coordinates in centers.items():
        sum_x = 0
        sum_y = 0
        count = 0
        for point, point_coordinates in points.items():
            if distance(point_coordinates, center_coordinates) == min(distances[point].values()):
                sum_x += point_coordinates[0]
                sum_y += point_coordinates[1]
                count += 1
        if count > 0:
            new_centers[center] = (sum_x / count, sum_y / count)
        else:
            new_centers[center] = center_coordinates
    return new_centers
```

```python
def print_points():
    # Вывод точек и центров с использованием matplotlib
    plt.figure(figsize=(15, 8))
    plt.scatter(*zip(*centers.values()), color='red', label='Центры1')
    plt.scatter(*zip(*points.values()), color='blue', label='Точки')

    # Добавление подписей для каждой точки и центра
    for label, (x, y) in {**centers, **points}.items():
        plt.annotate(label, (x, y), textcoords="offset points", xytext=(0, 10), ha='center')

    plt.legend()
    plt.show()
```

```python
def calculate_data():
    #Вычисление расстояний каждой точки до каждого центра
    distances = get_distances()
    # Вывод результатов
    for point, point_distances in distances.items():
        print(f"Расстояния для точки {point}:")
        for center, distance in point_distances.items():
            print(f"До центра {center}: {distance}")
        print()

    #Расчет суммарной ошибки
    error = sum(min(distances[point].values()) ** 2 for point in points)
    print(f'Суммарная ошибка = {error}')

    # Вычисление новых координат центров
    global centers
    centers = calculate_new_centers()
    print("Новые координаты центров:")
    for center, center_coordinates in centers.items():
        print(f"{center}: {center_coordinates}")
```

```python
if __name__ == '__main__':
    cents = centers
    old_cents = {}
    print_points()
    while old_cents != cents:
        calculate_data()
        print_points()
        old_cents = cents
        cents = centers

    print("Центры остались неизменными, выполнение завершено!")
    print_points()
```
